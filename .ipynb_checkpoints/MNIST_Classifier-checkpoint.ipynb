## Normalize the input

In order to work with this data it is useful to first normalize the input to the interval between 0 and 1:

```python
train_images = train_images / 255.0

test_images = test_images / 255.0
test_images.shape
```

# Building the model

We utilise the Keras framework to construct a network that can be trained to recognise the images. A high-level interface for describing neural networks is called Keras. Tensorflow comes with one implementation of Keras, that we are using here.

Creating a model that arranges the varioulayersrs of our network is the first step. The layers in our scenario will be layered sequentially, and we'll be using tf.keras.Sequential:


```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

The input layer of the model creates a 1D array by flattening the 28*28 pixel pictures. The ReLU activation function in the second layer, which is the hidden layer and includes 128 neurons, adds non-linearity to the model and aids in its ability to learn complex patterns. Ten neurons in the output layer each represent a class label (0 to 9) for each digit. The softmax parameter defines the softmax activation function, which transforms the model's raw output into probabilities, each of which represents the likelihood that an input would fall into a specific class.


```python
model.summary()
```


Here, we can observe that {(None,784)} is the *flatten* layer's output form. In this case, $784 = 28\times28$ and {None} stand for arbitrary, undefined dimensions. This layer has no training parameters. Since there are 128 nodes in the first dense layer, the output shape is {(None, 128)}.  The weights $w^1_{i,j}$ that join the 128 inputs of the current layer with the 784 outputs of the preceding layers make up the number of parameters. Furthermore, there is one bias parameter $b^1_i$ for each of the 128 nodes. This indicates that there are a total of $(784+1) \times 128=100480$ parameters that we can change in this layer.  Likewise, there are $(128+1)\times 10 = 1290$ parameters to adjust in the final layer.

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

The optimizer, loss function, and evaluation metrics are specified while configuring a neural network's training process using Keras' `compile` method. `Adam` is a popular choice for the `optimizer`, which controls how the model's parameters are changed during training. For classification problems, the loss function measures the difference between the true and predicted labels, e.g., 'sparse_categorical_crossentropy'. Metrics such as 'accuracy' evaluate the performance of the model during training. The compile technique modifies the model's learning process by adjusting these parameters, which impacts the model's capacity to converge on optimal solutions and produce precise predictions on unobserved data.

## Training the Network

Finally we are now in a position to train our network.  This is done through the `fit()` method:

```python
model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print('Test accuracy:', test_acc)
```

We can see from the evaluation on test data, the accuracy has come out to be very good as 97% with minimal loss of 0.07. The model is considered good here.

```python
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array, true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array, true_label[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

The code above provides two functions for plotting images and their corresponding prediction confidence values. Plot image function displays the image with its predicted label and confidence percentage. If prediction matches the true label, the label is displayed in blue, otherwise in red. The plot value array function visualizes the prediction confidence values as a bar chart, with each bar representing a class.  

```python
i = 200
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i],  test_labels)
plt.show()
```

```python
offset=20
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i+offset, predictions[i+offset], test_labels, test_images)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i+offset, predictions[i+offset], test_labels)
plt.tight_layout()
plt.show()
```

This code generates a grid of subplots to visualize a set of images along with their predicted labels and confidence values. It calculates the total number of images to display based on the number of rows and columns specified. It then iterates through each subplot, plotting the image and its corresponding prediction. The plot_image function displays the image with its predicted label and color-coded true label. The plot_value_array function visualizes the prediction confidence values as a bar chart. The offset variable allows starting the visualization from a specific index in the dataset. The final layout is adjusted for better spacing between subplots using plt.tight_layout(), and the plot is displayed using plt.show().

## üß™ Hyperparameter Tuning & Performance Analysis

To improve the performance of the digit classification model, several experiments were conducted by adjusting key hyperparameters. Here's a summary of the changes, their purpose, and observed impact:

---

### üîº Increased Number of Nodes
The number of neurons in the dense (fully connected) layer was increased from **128 to 256**. This allowed the model to learn more complex and abstract patterns from the training data by capturing subtle features that may have been missed with fewer nodes.

‚úÖ **Effect:** Improved the model's capacity and slightly boosted accuracy on the test set.

---

### üßØ Added Dropout Layer
A **Dropout layer with a rate of 0.2** was added after the dense layer. Dropout works by randomly setting a fraction of input units to zero during training, which helps prevent overfitting by encouraging the model to learn more robust and generalizable features.

‚úÖ **Effect:** Helped reduce overfitting and slightly improved test accuracy.

---

### üîÅ Increased Number of Epochs
Training epochs were increased from **5 to 10**, giving the model more passes over the training data. This refinement helped the model better adjust its weights and generalize more effectively.

‚úÖ **Effect:** Improved test accuracy without overfitting.

---

### üì¶ Batch Size Tuning
Different batch sizes were tested. While **larger batch sizes** can speed up training, they may lead to less stable updates. Smaller batch sizes offer more stable convergence but take longer.

After testing, a **batch size of 32** was retained because:
- It provided the best balance between training time and performance.
- Increasing it did **not result in significant accuracy gains**.

‚úÖ **Effect:** Maintained strong accuracy; avoided unnecessary trade-offs.

---

### ‚ùå Changing the Loss Function
The loss function was changed to **`mean_squared_error`**, which is typically used in regression tasks. As expected, this led to **very poor accuracy** on the test set.

‚úÖ **Conclusion:** For multi-class classification problems like digit recognition, **`sparse_categorical_crossentropy`** (or `categorical_crossentropy` if labels are one-hot encoded) is more appropriate.

---

### ‚úÖ Summary of Impact
After tuning, the improved model reached **98% accuracy** on the test set. These changes demonstrate the importance of:
- Matching hyperparameters to the task
- Balancing model complexity with regularization
- Evaluating both performance and training efficiency





```python
model1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  
    tf.keras.layers.Dense(256, activation=tf.nn.relu), #changing the nodes in the dense layer
    tf.keras.layers.Dropout(0.2), #adding the dropout layer 
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

model1.summary()

model1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])  
model1.fit(train_images, train_labels, epochs=10, batch_size = 32) # changing the epochs 

```

```python
test_loss, test_acc = model1.evaluate(test_images, test_labels)

print('Test accuracy:', test_acc)
```

## ‚ùó Model Limitations & Generalization Test

Although the trained model achieved a high accuracy of **97%** on the MNIST test dataset, it failed to correctly classify some **custom or external test images** that were provided manually.

This highlights an important limitation:

> ‚ö†Ô∏è The model performs well on data that is similar in style and format to the original MNIST dataset, but struggles when presented with images that differ in nature, appearance, or noise.

**Reason:**  
The MNIST dataset contains grayscale, centered, and well-aligned handwritten digits on a uniform background. If new test images differ in terms of:
- background noise
- lighting conditions
- digit thickness or style
- positioning and alignment

‚Ä¶ then the model may fail to generalize well.

### üß† Key Insight
Even with high test accuracy, a model can struggle with **distribution shift** or **out-of-domain samples**. This suggests the importance of:
- Data augmentation during training
- More diverse training data
- Using more robust models (e.g., deeper CNNs or pretrained vision models for broader generalization)
